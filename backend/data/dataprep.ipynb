```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
orders=pd.read_csv("./olist_orders_dataset.csv")
payments=pd.read_csv("./olist_order_payments_dataset.csv")
```

```python
orders.head()
```

```python
payments.head()
```

```python
df = pd.merge(orders, payments, on="order_id", how="inner")
df
```

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df = df.sort_values('order_purchase_timestamp')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['order_approved_at','order_delivered_carrier_date','order_delivered_customer_date'],inplace=True)
df.isna().sum()
```

```python
final_df = df[['order_id', 'order_status', 'payment_value']]
```

```python
final_df.to_csv("../public/data.csv", index=False)
```
